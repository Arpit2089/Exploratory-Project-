```python
import numpy as np
import random

destination="240750082407505"
population_size=20
Gene_pool="abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"
population=["".join(random.choices(Gene_pool, k=15)) for _ in range(population_size)]

def objective(x):
  fitness=0
  for i in range(len(x)):
    if x[i]==destination[i]:
        fitness-=1
  return fitness

def crossover(parent1,parent2):
  x=np.random.randint(0,16)
  child1=parent1[:x]+parent2[x:]
  child2=parent2[:x]+parent1[x:]
  return [child1,child2]

hyperparameters=[5]
max_epochs=2500
while max_epochs:
  population.sort(key=lambda x: objective(x))
  # print(population[0])
  selected_population=population[:hyperparameters[0]]
  population=[]
  population_size=20
  while population_size:
    population_size-=2
    parent1=selected_population[np.random.randint(0,len(selected_population))]
    parent2=selected_population[np.random.randint(0,len(selected_population))]
    child1,child2=crossover(parent1,parent2)

    if np.random.rand()<0.05 :
        mutation_point1=np.random.randint(0,15)
        new_part=random.choices(Gene_pool,k=1)
        child1=child1[:mutation_point1]+random.choice(Gene_pool)+child1[mutation_point1+1:]
    if np.random.rand()<0.05 :
        mutation_point2=np.random.randint(0,15)
        child2=child2[:mutation_point2]+random.choice(Gene_pool)+child2[mutation_point2+1:]

    population.append(child1)
    population.append(child2)
  max_epochs-=1
population.sort(key=lambda x: objective(x))
print(population[0])








```
